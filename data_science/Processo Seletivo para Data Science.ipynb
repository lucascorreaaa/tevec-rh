#### 7. Explique para um leigo, como funciona um classificador RandomForest.

Antes de explicar o **funcionamento** acho importante introduzir rapidamente **o que é** a técnica de Random Forest e um classificador Random Forest.

**Random Forest (Floresta Aleatória)** é uma técnica de Aprendizado de Máquina (sub-área do campo de Inteligência Artificial em computação) que possui duas capacidades:
* **Regressão (estimar valores contínuos)**: estimar o preço de um imóvel, a temperatura de um ambiente, o valor de uma ação na bolsa, etc..
* **Classificação (identificar categorias/classes)**: identificar se uma transação é ou não fraudulenta, se um email é ou não spam. Serve também para identificar e classificar objetos em imagens, como animais, frutas, veículos, pessoas, etc..
Essa técnica é capaz de fazer essas estimativas e classificações a partir de um conjunto de exemplos e de características do objeto observado.

A técnica de Random Forest que aplica a **classificação** é conhecida, na verdade, como ***Random Forest Classifier* (RFC) (Classificador de Floresta Aleatória)**.

Para entender o funcionamento dos RFCs é fundamental que entendamos o que é uma **Árvore de Decisão**, isto porque a técnica de Floresta Aleatória tem o termo "Floresta" em seu título não por acaso mas por utilizar-se de múltiplas árvores de decisão em seu processo de classificação.

Uma **Árvore de Decisão (*Decision Tree*)** é um estrutura hierárquica que busca identificar classes a partir de características do objeto observado. Para entender melhor, vamos estabelecer um exemplo real: 

Assuma a tabela abaixo como nosso conjunto de dados, contendo 5 exemplos de frutas com as seguintes características: cor, tamanho(cm) e nome. Uma árvore de decisão que nos ajude a identificar a fruta (nome) a partir de suas características (cor e tamanho, neste exemplo) utilizará os 5 exemplos de frutas que temos a disposição para descobrir quais características são mais relevantes para o processo de identificação de uma determinada fruta. 

| Id | Cor      | Tamanho (cm)   | Nome          |
|----|----------|----------------|---------------|
| 1  | verde    | 20             | coco verde    |
| 2  | verde    | 32             | melancia      |
| 3  | amarelo  | 18             | melão         |
| 4  | amarelo  | 9              | maracujá doce |
| 5  | vermelho | 6              | maçã          |

Utilizando os 5 exemplos acima, uma árvore de decisão poderia atribuir ao seu **nó raiz (o primeiro nó da árvore)** o atributo "cor". Isso significa que a partir deste nó raiz, porque ele representa o atributo "cor", se particionará os dados em 3 partes (3 nós, descendentes do nó raiz), cada uma delas contendo todos os registros que possuem um valor específico do atributo "cor". Ilustrei a arquitetura da árvore até este momento na imagem abaixo. Os números nos nós correspondem aos "Ids" dos registros de acordo com nossa tabela (conjunto de dados).

![arvore1](img/img1.png)

Observe que, dado este nosso conjunto restrito de 5 exemplos de frutas, com a informação de que a cor da fruta é vermelha eu consigo afirmar com certeza de que se trata de uma maçã, isso porque a maçã (Id 5 na árvore) é a única fruta do subconjunto das frutas vermelhas. No entanto, se eu tiver a informação de que a fruta é verde ou amarela, eu não consigo afirmar com certeza de qual fruta se trata, porque tanto no subconjunto das frutas verdes quanto no das frutas amarelas há mais de uma fruta.

O objetivo da árvore de decisão é ir criando esses subconjuntos de dados a partir dos atributos dos nossos exemplos, até que em todos os subconjuntos finais da árvore existam apenas exemplos que pertençam a mesma classe. Neste exemplo das frutas temos 5 classes, pois temos na tabela 5 frutas diferentes (coco verde, melancia, melão, maracujá doce e maçã). Em uma aplicação em que a árvore de decisão é responsável por identificar se um email é considerado SPAM ou não, o número de classes é 2 (SPAM, não-SPAM). Numa aplicação anti-fraude, o número de classes também é 2 (fraude, não-fraude).

Voltando ao exemplo; observe na imagem que dos 3 subconjuntos que temos até agora na árvore, apenas o terceiro possui unicamente elementos da mesma classe (maçã), os outros dois possuem registros com classes distintas entre eles. Para resolver isso precisamos selecionar outra característica, ou seja, outro atributo do nosso conjunto de dados que nos ajude a particionar esses registros de modo que cada um deles fique em um novo subconjunto onde todos os elementos possuam a mesma classe que a sua. Como já utilizamos o atributo "cor", resta-nos apenas o atributo "tamanho". Podemos ver na imagem abaixo a arquitetura da árvore após a inclusão desse atributo como nó descendente do nó raiz (cor).

![arvore2](img/img2.png)

Utilizando diferentes valores de tamanho como condição para particionar os registros conseguimos obter 4 novos subconjuntos, cada um contendo apenas elementos de uma mesma classe, que era nosso objetivo inicial. Pra este conjunto de dados das frutas que especificamos, em particular, cada subconjunto possui apenas 1 registro, mas poderia ter vários.

Com essa arquitetura da árvore de decisão bem definida podemos então utilizá-la para classificar novos registros de frutas desconhecidas, desde que tenhamos dados sobre os outros atributos como cor e tamanho. Por exemplo, se quisermos saber de que fruta se trata o registro abaixo, basta processarmos seus atributos na árvore de decisão.

| Id | Cor   | Tamanho (cm) | Nome |
|----|-------|--------------|------|
| 6  | verde | 22           | ??   |

No nó raiz, do atributo "cor", o registro é incluído no subconjunto das frutas verdes. No nó seguinte, do atributo "tamanho", o registro é direcionado ao nó folha que contém o subconjunto das frutas de cor verde e de tamanho menor que 25cm. O **nó folha** representa o último nível para aquela ramificação da árvore de decisão, ou seja, todos os elementos de seu subconjunto pertencem a mesma classe. Portanto, como nosso novo registro atingiu o nó folha da classe "coco verde", classificamos o registro 6, até então desconhecido, como sendo um "coco verde", como ilustrado na imagem abaixo.

![arvore3](img/img3.png) 

Com o conceito de Árvore de Decisão esclarecido, podemos finalmente explicar como funciona o Classificador Random Forest (RFC). Como já adiantei anteriormente, **o RFC se utiliza de um conjunto de árvores de decisão em seu processo de classificação**. Porém, essas árvores não utilizam todos os dados disponíveis na base e nem todos os atributos. Em vez disso, o algoritmo prevê uma **amostragem aleatória (*random*) tanto de registros quanto de atributos/características** para ser trabalhada em cada árvore de decisão da floresta formando sua arquitetura. Esse processo é ilustrado na imagem abaixo.

![arvore4](img/img4.png)

Nesse exemplo temos um RFC composto de **oito árvores de decisão**, cada uma com sua arquitetura própria. Vamos estender nosso exemplo da árvore de decisão classificadora de frutas para enteder o funcionamento do RFC. Em vez de apenas 5 registros e 3 atributos (cor, tamanho e nome), imagine que nossa base de dados agora possui 10.000 registros e 15 atributos. No processo de escolha aleatória, **parcelas desses registros e atributos irão compôr cada uma das amostras**, que por sua vez alimentarão as 8 árvores de decisão para definirem suas arquiteturas.

Após as arquiteturas das árvores de decisão estarem formadas podemos tentar classificar um novo registro de fruta utilizando o RFC. Os atributos do registro passarão por todas as 8 árvores de decisão da floresta. Cada uma delas classificará o novo registro de acordo com sua arquitetura. Devido ao fato das amostras que formaram as árvores de decisão serem diferentes, as arquiteturas das árvores também são diferentes umas das outras. O efeito disso é que um mesmo registro pode ser atribuído a diferentes classes no conjunto de árvores da floresta (e isso é perfeitamente normal). A imagem abaixo ilustra esse cenário.

![arvore5](img/img5.png)

**O trabalho do RFC é justamente utilizar as respostas de todas as árvores da floresta para estabelecer um consenso e definir de fato a qual classe o novo registro pertence**. Isso é feito de maneira bem democrática; cada árvore de decisão tem seu poder de voto. A classe que for "mais votada", isto é, **a classe que tenha sido o resultado mais frequente entre todas as árvores de decisão da floresta ganha a disputa e tem o novo registro associado a si pelo RFC**. Na floresta da imagem acima podemos observar que a maior ocorrência (5 de 8, 62.5%) entre as árvores é da classe **"melancia"**, portanto, essa seria a resposta do RFC sobre a classe do novo registro.

Imagine que temos dados de peso, altura e idade de 20 pessoas diferentes, conforme a tabela acima. Considere **"idade"** nossa variável de interesse a ser estimada em novos registros, ou seja, nossa variável **Y**. As outras variáveis, "altura" e "peso", fazem parte do nosso conjunto de características **X**. 

O modelo K-NN fará exatamente seu papel de regressor, conforme definimos anteriormente, isto é, se baseará nesses 20 exemplos para estimar a idade de novas pessoas a partir de seus dados do conjunto de característica(altura e peso, neste caso).

A sigla K-NN é da expressão em inglês **"*K-Nearest Neighbours*"**, que em português significa **"K vizinhos mais próximos"**, e este nome tem tudo a ver com o funcionamento desse método. O K-NN parte do princípio que **registros que possuem valores parecidos de Y (variável a ser estimada) tendem a ter valores de seu conjunto de características X também parecidos**. Portanto, para estimar o valor da variável de interesse de um novo registro, o modelo K-NN visita todos os registros de sua base e busca pelos **K registros que possuem valores do conjunto X mais parecidos com os do novo registro**. K é um número a ser definido pela pessoa que implementa o modelo K-NN, ele determina **quantos vizinhos mais próximos daquele novo registro** o algoritmo irá considerar para fazer sua estimativa final, que consiste simplesmente em **tirar a média entre os valores da variável de interesse Y de cada um desses vizinhos**.

### 1.2 - Matriz de Correlação e Extração de Features

Abaixo geramos um mapa de calor (*heatmap*) para visualizarmos o nível de correlação entre as variáveis. É possível observer que algumas delas possuem alta correlação com outras.

```python
corr = X.corr()
sns.heatmap(corr, vmin=-1, vmax=1, center=0, cmap=sns.diverging_palette(20, 220, n=200))
plt.show()
```

No código abaixo estabelecemos o **limite do valor de correlação** em **0.8**. As variáveis que se correlacionam com outras com fator absoluto maior que 0.8 são excluídas (entre duas variávveis cujo fator de correlação ultrapassa o limite estabelecido a escolhida para ser excluída é a que possui o maior índice dentro do conjunto de variáveis, isso porque a exclusão é feita baseada no triângulo superior da matriz de correlação). Vamos visualizar a seguir quais variáveis serão excluídas.

```python
# Gettin upper triangle of correlation matrix
corr_upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(np.bool))

# Identifying features (columns) to drop (corr value > 0.8)
columns_to_drop = [column for column in corr_upper.columns if any(corr_upper[column].abs() > 0.8)]
print(columns_to_drop)
```

Em seguida excluímos do nosso conjunto de exemplos X as variáveis acima identificadas, e geramos um novo mapa de calor apenas para visualizarmos o desaparecimento das células mais escuras.

```python
# Droping columns from X
X = X.drop(columns_to_drop, axis=1)

# Plotting the heatmap of correlation matrix without correlated features.
sns.heatmap(X.corr(), vmin=-1, vmax=1, center=0, cmap=sns.diverging_palette(20, 220, n=200))
plt.show()
```

### 1.3 - Redução de Dimensionalidade com PCA

Com a análise da matriz de correlação pudemos excluir **10 variáveis** que possuíam alta correlação com outras, restando 90 variáveis, um número ainda muito grande. Seria bom se pudéssemos reduzir ainda mais a dimensionalidade dos nossos dados. Para isso, vamos utilizar a técnica de **Análise de Componentes Principais (*Principal Component Analysis - PCA*)**.

Nosso *threshold* para a expressividade da variância será de **0.999**, ou seja, devemos selecionar as **N componentes** com maior autovalor que em conjunto representem **99,9%** da variância dos dados. Abaixo podemos ver que **40 componentes** foram mantidos da PCA. Nosso conjunto de dados teve uma redução de dimensionalidade de **55,55%**, um grande avanço, conseguimos nos livrar das variáveis irrelevantes mencionadas no enunciado.

```python
# Applying PCA over all features and getting sufficient components to explain 99,9% of variance
pca = PCA(n_components=0.999, svd_solver='full')

# Fitting PCA and transforming X set onto PCA components space (dimensionality reduction)
X_pca = pca.fit_transform(X)
print(len(pca.components_))
```

### 1.4 - Treinando o classificador Extra Trees

Terminada a parte de *data preparation* vamos enfim treinar nosso modelo de classificação.

Primeiramente, à partir do nosso conjunto de dados *X_pca*, criamos os **conjuntos de treinamento** (*X_train*, *y_train*) e de **teste** (*X_test*, *y_test*). Em seguida, treinamos nosso modelo de classificação ***Extra Trees - Extremely Randomized Trees Classifier* (ETC)** utilizando **100 estimadores** (árvores de decisão aleatórias) sobre nosso conjunto de treinamento (*X_train*, *y_train*).

Por fim, utilizamos o modelo treinado para fazer a classificação do conjunto de teste *X_test*, guardamos o resultado na variável *y_pred* e comparamos com o conjunto de teste *y_test* da variável dependente para sabermos a performance do modelo. 

Podemos ver abaixo que a performance do nosso classificador foi de **81,00%**.

```python
# Spliting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2)

# Fitting RFC model
etc = ExtraTreesClassifier(n_estimators=100)
etc.fit(X_train, y_train)

# Classifing the test set
y_pred = etc.predict(X_test)

# Checking RFC model performance
rfc_performance = (y_pred == y_test).sum() / len(y_test)
print("Extra Trees Classifier performance:", rfc_performance*100, "%")
```

### 1.5 - Por que Extra Trees?

O classificador ***Extra Trees*** é um método de *ensemble* que funciona de maneira semelhante ao modelo de *Random Forest*, com a importante diferença de utilizar em sua floresta **árvores de decisão extremamente aleatórias** (*Extremely Randomized Trees*). Essas árvores são também muito parecidas com as árvores de decisão tradicionais, mas possui uma diferença crucial: a seleção dos valores de divisão dos atributos numéricos, que estão sob análise em cada nó, não é feita de maneira a minimizar a entropia, mas sim de maneira completamente aleatória. Caso o atributo seja categórico o ganho de entropia é igualmente ignorado e a categoria é escolhida de maneira aleatória. Isso torna a árvore menos enviesada pelas *features* do conjunto de dados. 

Essa pequena mas significativa diferença faz com que o classificador *Extra Trees* consiga diminuir a variância e eventualmente aumentar, de maneira não significativa, o *bias*. Pelo modo de funcionamento particular de suas árvores de decisão extremamente aleatórias, o classificador ***Extra Trees* normalmente performa melhor que o *Random Forest* quando o conjunto de dados inclui muitas variáveis numéricas, assim como o nosso**, o que justifica minha opção por esse classificador.

### Item 3 - Solução

Eu **fiz toda a solução na própria classe ATM** do arquivo **caixa_eletronico.py**, utilizando **2 métodos**: **process_request()** e **output_message()**. Para testar a solução basta rodar o arquivo **caixa_eletronico.py**, mas também coloquei abaixo a classe ATM com os dois métodos criados, para visualização.

```python
class ATM():
    import numpy as np
    import json

    def process_request(self, input):

        # Calculating minimum required quantity
        amount = input['request']['request']
        user_id = input['request']['userID']
        n_50,rem_50 = np.divmod(amount,50)
        n_20,rem_20 = np.divmod(rem_50,20)
        
        # Returning output SUCCESS message
        if rem_20 == 0:
            return self.output_message(user_id, amount, n_20, n_50)

        # Returning output SUCCESS message after adjust the number of notes (case of 80 and 60 decimals)
        elif rem_20 == 10:
            n_50 -= 1
            n_20_incrementer = 3 if n_20 == 0 else 4
            n_20 += n_20_incrementer
            return self.output_message(user_id, amount, n_20, n_50)
        # Returning output FAIL message 
        else:
            return self.output_message(user_id, amount, fail=True)

    def output_message(self,user_id, requested_amount, n_20, n_50, fail=False):
        if fail:
            message = {
                        'requester': {
                            'userID': user_id,
                            'requested': requested_amount
                        },
                        'response': {
                            'error': 'Nao e possivel sacar a quantia solicitada porque nao ha notas compatives. As notas disponiveis sao de R$ 20 e R$ 50, por favor insira um valor compativel.'
                        }
                    }
            return message
        message = {
                    'requester': {
                        'userID': user_id,
                        'requested': requested_amount
                    },
                    'response': {
                        '20': n_20.item(),
                        '50': n_50.item()
                    }
                }
        return message
```
